```python
import numpy as np

def get_direction(entry_direction, field):
    
    if field == 'S': return (0,0)
    
    if [entry_direction, field] in [[(-1,0), '|'], [(0,-1), 'L'], [(0,1), 'J']] : return (-1,0)
    if [entry_direction, field] in [[(-1,0), 'F'], [(0,1), '-'], [(1,0), 'L']] : return (0,1)
    if [entry_direction, field] in [[(1,0), '|'], [(0,1), '7'], [(0,-1), 'F']] : return (1,0)
    if [entry_direction, field] in [[(-1,0), '7'], [(1,0), 'J'], [(0,-1), '-']] : return (0,-1)
    
    return False
    
def walk(matrix, pos, direction, steps = 0):
    
    def next_step_possible(matrix, pos, direction):
        
        # possible next tube
        pos_next_pos = np.add(pos, direction)
        
        # check if next field would be in range
        if not (0 <= pos_next_pos[0] < matrix.shape[0] and 0 <= pos_next_pos[1] < matrix.shape[1]):
            return False, None, None
        
        next_coordinates = tuple(pos_next_pos)
        
        # check if next field can be reached
        next_direction = get_direction(direction, matrix[next_coordinates])
        if next_direction is False:
            return False, None, None
        
        return True, next_direction, next_coordinates
        
    # check if a step is possible
    next_step_is_possible, next_direction, next_coordinates = next_step_possible(matrix, pos, direction)
    
    while next_step_is_possible:
        
        # make step 
        pos = next_coordinates
        direction = next_direction
        steps += 1
        
        # next step is 'S'
        if direction == (0,0):
            return True, steps
        
        next_step_is_possible, next_direction, next_coordinates = next_step_possible(matrix, pos, direction)
        
    return False, None
```

```python
def solve_puzzle(puzzle_input):
    
    def convert_input_to_matrix(input_data):
        
        return np.array([[char for char in line] for line in input_data.splitlines()])
        
    # movements: north, south, west, east
    movements_list = [(-1,0), (1,0), (0,-1), (0,1)]

    # convert input to matrix
    map_matrix = convert_input_to_matrix(puzzle_input)

    # finde start
    start_cords_tupel = next(indexes for indexes, cell in np.ndenumerate(map_matrix) if cell == 'S')
    steps = 0
    
    for direction in movements_list:
        
        possible, steps_to_start = walk(map_matrix, start_cords_tupel, direction, steps)
        
        if possible:
            steps = steps_to_start
            break
    
    return int(steps / 2)
            
solve_puzzle(puzzle_input)
```
