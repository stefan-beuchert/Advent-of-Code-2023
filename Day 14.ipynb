```python
import numpy as np

def transform_input_to_matrix(data):
    
    return np.array([[char for char in line] for line in data.splitlines()])

def rolle_north(matrix):
    
    def rolle_west(list_of_rocks):
    
        left_pointer = 0

        for right_pointer in range(len(list_of_rocks)):

            # if no stone found push left pointer
            if list_of_rocks[right_pointer] == '.':
                continue

            # if square stone found, set left pointer to the position behinde that
            elif list_of_rocks[right_pointer] == '#':
                if right_pointer + 1 < len(list_of_rocks):
                    left_pointer = right_pointer + 1

            # if round stone found, swap place with left pointer
            elif list_of_rocks[right_pointer] == 'O':
                list_of_rocks[right_pointer] = '.'
                list_of_rocks[left_pointer] = 'O'

                left_pointer += 1

        return list_of_rocks
    
    matrix_t = matrix.transpose()
    
    for row_index in range(len(matrix_t)):
        
        matrix_t[row_index] = rolle_west(matrix_t[row_index])
        
    return matrix_t.transpose()

def get_weight(matrix):
    
    res = 0
    
    for index, row in enumerate(matrix):
        res += sum(element == 'O' for element in row) * (len(matrix) - index)
    
    return res
```

```python
def solve_puzzle(input_data):
    
    matrix = transform_input_to_matrix(input_data)
    stones_shifted_matrix = rolle_north(matrix)
    
    return get_weight(stones_shifted_matrix)
    
solve_puzzle(puzzle_data)
```
