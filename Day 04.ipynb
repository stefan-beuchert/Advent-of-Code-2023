```python
class Card:
    
    def __init__(self, card_id, card_numbers_as_string):
        self.id = card_id
        
        # extract winning numbers and scratched numbers
        number_groups_in_string = card_numbers_as_string.split('| ')
        
        self.list_of_wining_numbers = self.extract_numbers_as_int_list(number_groups_in_string[0])
        self.list_of_scrateched_numbers = self.extract_numbers_as_int_list(number_groups_in_string[1])
        
    def __str__(self):
        print(f'id: {self.id} \nwining numbser: {self.list_of_wining_numbers} \nscrateched numbers: {self.list_of_scrateched_numbers}')
        
    def extract_numbers_as_int_list(self, string_of_numbers_separated_by_whitspaces):
        
        list_of_strings = string_of_numbers_separated_by_whitspaces.split(' ')
        list_of_strings_without_whitspaces = [element for element in list_of_strings if element != '']
        list_of_ints = [int(element) for element in list_of_strings_without_whitspaces]
        
        # sort list
        list_of_ints.sort()
        
        return list_of_ints
    
    def get_points(self):
        
        # finde number of ints in self.scratched_numbers that is also in self.wining_numbers
        numbers_in_both_lists = set(self.list_of_wining_numbers).intersection(set(self.list_of_scrateched_numbers))
        list_of_numbers_in_both_lists = list(numbers_in_both_lists)
        
        number_of_wins = len(list_of_numbers_in_both_lists)
        points = 2**(number_of_wins-1)
        
        return int(points)
    
    def get_number_of_new_cards(self):
        # finde number of ints in self.scratched_numbers that is also in self.wining_numbers
        numbers_in_both_lists = set(self.list_of_wining_numbers).intersection(set(self.list_of_scrateched_numbers))
        list_of_numbers_in_both_lists = list(numbers_in_both_lists)
        
        number_of_wins = len(list_of_numbers_in_both_lists)
        
        return number_of_wins
```

```python
# solve puzzle

def solve_puzzle(puzzle_input):
    list_of_cards_as_string = puzzle_input.split('\n')
    
    points = []
    
    for index in range(len(list_of_cards_as_string)):
        
        card_as_string = list_of_cards_as_string[index]
        card_in_two_parts = card_as_string.split(': ')
        card_numbers_as_string = card_in_two_parts[1]
        
        card = Card(index + 1, card_numbers_as_string)
        
        points.append(card.get_points())
    
    res = sum(points)
    
    return res
```

--- Part Two ---
Just as you're about to report your findings to the Elf, one of you realizes that the rules have actually been printed on the back of every card this whole time.

There's no such thing as "points". Instead, scratchcards only cause you to win more scratchcards equal to the number of winning numbers you have.

Specifically, you win copies of the scratchcards below the winning card equal to the number of matches. So, if card 10 were to have 5 matching numbers, you would win one copy each of cards 11, 12, 13, 14, and 15.

Copies of scratchcards are scored like normal scratchcards and have the same card number as the card they copied. So, if you win a copy of card 10 and it has 5 matching numbers, it would then win a copy of the same cards that the original card 10 won: cards 11, 12, 13, 14, and 15. This process repeats until none of the copies cause you to win any more cards. (Cards will never make you copy a card past the end of the table.)

This time, the above example goes differently:

Card 1: 41 48 83 86 17 | 83 86  6 31 17  9 48 53

Card 5: 87 83 26 28 32 | 88 30 70 12 93 22 82 36
Card 6: 31 18 13 56 72 | 74 77 10 23 35 67 36 11
Card 1 has four matching numbers, so you win one copy each of the next four cards: cards 2, 3, 4, and 5.
Your original card 2 has two matching numbers, so you win one copy each of cards 3 and 4.
Your copy of card 2 also wins one copy each of cards 3 and 4.
Your four instances of card 3 (one original and three copies) have two matching numbers, so you win four copies each of cards 4 and 5.
Your eight instances of card 4 (one original and seven copies) have one matching number, so you win eight copies of card 5.
Your fourteen instances of card 5 (one original and thirteen copies) have no matching numbers and win no more cards.
Your one instance of card 6 (one original) has no matching numbers and wins no more cards.
Once all of the originals and copies have been processed, you end up with 1 instance of card 1, 2 instances of card 2, 4 instances of card 3, 8 instances of card 4, 14 instances of card 5, and 1 instance of card 6. In total, this example pile of scratchcards causes you to ultimately have 30 scratchcards!

Process all of the original and copied scratchcards until no more scratchcards are won. Including the original set of scratchcards, how many total scratchcards do you end up with?

```python
# solve the second part
def solve_second_part(puzzle_input):
    
    # convert input to list of cards
    list_of_cards_as_string = puzzle_input.split('\n')
    
    list_of_cards = []
    
    for index in range(len(list_of_cards_as_string)):
        
        card_as_string = list_of_cards_as_string[index]
        card_in_two_parts = card_as_string.split(': ')
        card_numbers_as_string = card_in_two_parts[1]
        
        card = Card(index + 1, card_numbers_as_string)
        
        list_of_cards.append(card)
        
    # create list to keep track on number of cards
    list_of_number_of_cards = [1]*len(list_of_cards)
    
    # calculate number of new cards for each card
    for index in range(len(list_of_cards)):
        
        # get number of current cards
        number_of_current_cards = list_of_number_of_cards[index]
        
        # get number of wins for this card
        number_of_new_cards = list_of_cards[index].get_number_of_new_cards()
            
        # update number of following cards
        for counter in range(number_of_new_cards):
            list_of_number_of_cards[index + counter + 1] = list_of_number_of_cards[index + counter + 1] + number_of_current_cards
            
    total_number_of_cards = sum(list_of_number_of_cards)
    
    return total_number_of_cards
```
