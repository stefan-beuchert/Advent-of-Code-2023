```python
import numpy as np

def get_empty_rows_indices(m, empty = '.'):
        
    empty_row_indexes = []

    for row_index in range(len(m)):
        if all(element == empty for element in m[row_index]):
            empty_row_indexes.append(row_index)

    return empty_row_indexes


def get_galaxy_coordinates(m, expansion_rate, empty_row_indices, empty_col_indices, galaxy = '#'):
    
    def new_coordinate(origin, expansion_rate, list_indexes_to_expand):
        return origin + (expansion_rate - 1) * sum(i < origin for i in list_indexes_to_expand)
        

    galaxie_coordinates_list = []

    for indices_tupel, element in np.ndenumerate(m):
        if element == galaxy:

            # get new coordinates by expanding the universe (with the previously extraced indexes of empty rows/cols)
            galaxy_row_coordinate = new_coordinate(indices_tupel[0], expansion_rate, empty_row_indices)
            galaxy_col_coordinate = new_coordinate(indices_tupel[1], expansion_rate, empty_col_indices)

            galaxie_coordinates_list.append((galaxy_row_coordinate, galaxy_col_coordinate))

    return galaxie_coordinates_list


def total_min_distance_between_all_galaxies(list_of_coordinates):
    
    def path_length_between_galaxies(coords_gal_a, coords_gal_b):
    
        return abs(coords_gal_a[0] - coords_gal_b[0]) + abs(coords_gal_a[1] - coords_gal_b[1])

    total_distance = 0

    # calculate min distance between each possible galaxy combination
    for gal_a_index in range(len(list_of_coordinates)):
        for gal_b_index in range(gal_a_index + 1, len(list_of_coordinates)):
            total_distance += path_length_between_galaxies(list_of_coordinates[gal_a_index], 
                                                         list_of_coordinates[gal_b_index])

    return total_distance

def solve_puzzle(puzzle_input, expansion_rate = 2):
    
    # get data as matrix 
    matrix = np.array([[char for char in line] for line in puzzle_input.splitlines()])
    
    # get indexes of empty rows and cols
    empty_row_indexes = get_empty_rows_indices(matrix)
    empty_col_indexes = get_empty_rows_indices(matrix.transpose())
    
    galaxie_coordinates_list = get_galaxy_coordinates(matrix, expansion_rate, 
                                                      empty_row_indexes, empty_col_indexes)
            
    return total_min_distance_between_all_galaxies(galaxie_coordinates_list)

solve_puzzle(puzzle_input)
```
