```python
# algorithm for solving the puzzle

def solve_calibration_problem(calibration_input):

    res_list_of_numbers = []

    # look at each line
    for line in calibration_input.splitlines():

        # extract first and last number
        first_number = None
        last_number = None

        # get first number
        for char in line:
            if char.isdigit():
                first_number = int(char)
                break

        # get second number (by itterating the reversed line)
        for char in line[::-1]:
            if char.isdigit():
                last_number = int(char)
                break

        # combine numbers to one integer
        line_value = first_number * 10 + last_number

        # save result in list
        res_list_of_numbers.append(line_value)

    # add all integers together
    final_res = sum(res_list_of_numbers)

    return final_res
```

```python
# algorithm for solving the puzzle
def alter_input_for_second_part(calibration_input):
    
    def check_if_number(string_with_possible_number):
        # the function takes a string and checks if it contains a one digit integer written out
        # one digit integers can be represented between 3 to 5 char strings (one, two, six and three, seven, eight)
        # therefor the input string will be cut to 5 chars and checked for each possible value for zero to nine
        # this function return the number as a singel char and the length of the number as string
        
        # check if char is even long enought to contain a number
        if len(string_with_possible_number) > 2:
        
            # check for special cases where numbers are intertwined
            if len(string_with_possible_number) > 4:
                char_with_len_five = string_with_possible_number[:5]

                if char_with_len_five == "twone": return ("21", 5)

            if len(string_with_possible_number) > 6:
                char_with_len_seven = string_with_possible_number[:7]

                if char_with_len_seven == "oneight": return ("18", 7)
                elif char_with_len_seven == "eightwo": return ("82", 7)

            if len(string_with_possible_number) > 7:
                char_with_len_eight = string_with_possible_number[:8]

                if char_with_len_eight == "fiveight": return ("58", 8)
                elif char_with_len_eight == "nineight": return ("98", 8)

            if len(string_with_possible_number) > 8:
                char_with_len_nine = string_with_possible_number[:9]

                if char_with_len_nine == "sevenine": return ("79", 9)

            if len(string_with_possible_number) > 9:
                char_with_len_ten = string_with_possible_number[:10]

                if char_with_len_ten == "threeight": return ("38", 10)

            # check if char is even long enought to contain a number
            if len(string_with_possible_number) > 2:
                char_with_len_three = string_with_possible_number[:3]

                if char_with_len_three == "one": return ("1", 3)
                elif char_with_len_three == "two": return ("2", 3)
                elif char_with_len_three == "six": return ("6", 3)

            # check if char is long enough for numbers with 4 chars
            if len(string_with_possible_number) > 3:
                char_with_len_four = string_with_possible_number[:4]

                if char_with_len_four == "four": return ("4", 4)
                elif char_with_len_four == "five": return ("5", 4)
                elif char_with_len_four == "nine": return ("9", 4)

            # check if char is long enough for numbers with 5 chars
            if len(string_with_possible_number) > 4:
                char_with_len_five = string_with_possible_number[:5]

                if char_with_len_five == "three": return ("3", 5)
                elif char_with_len_five == "seven": return ("7", 5)
                elif char_with_len_five == "eight": return ("8", 5)
            
        return None
    
    def replace_char_in_string(line, number_as_singel_char, length_of_number_as_char, index_of_first_letter_of_number):

        left = line[:index_of_first_letter_of_number]
        mid = number_as_singel_char
        end = line[index_of_first_letter_of_number + length_of_number_as_char:]

        new_line = left + mid + end
        
        return new_line
    
    
    new_input_list = []
    
    # go through each line
    for line in calibration_input.splitlines():
        
        # variables to calculate remaining chars in the while loop
        length_of_line = len(line)
        index = 0
        remaining_length_of_line = length_of_line - index
        
        # for each index in line
        while index < length_of_line:
            
            # check if current char is beginning of a number
            possible_number_as_char_and_length = check_if_number(line[index:])
            
            # if yes, replace string number with singel char number in line string
            if possible_number_as_char_and_length is not None:
                line = replace_char_in_string(line, possible_number_as_char_and_length[0], 
                                                  possible_number_as_char_and_length[1], index)
                
                # update variables for new line
                length_of_line = len(line)
            
            index = index + 1
            remaining_length_of_line = length_of_line - index
                                              
        new_input_list.append(line)
        
        # convert list to sting with line breaks
        new_input = '\n'.join(new_input_list)
        
    return new_input
```

```python
def solve_calibration_problem_part_two(calibration_input):
    
    # alter input
    new_input = alter_input_for_second_part(calibration_input)
    
    # solve with original function
    res = solve_calibration_problem(new_input)
    
    return res
```
