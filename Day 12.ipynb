```python
HISTORY_DICT = {}
```

```python
def get_combinations(list_of_springs, records_list):
    # this method takes an input string striped of all definet '.' and the record list
    
    if (list_of_springs, str(records_list)) in HISTORY_DICT:
        return HISTORY_DICT[(list_of_springs, str(records_list))]
    
    print("get combinations: " + str([list_of_springs, records_list]))
    
    number_of_springs = len(list_of_springs)
    total_number_of_combinations = 0
    
    # empty spring list
    if list_of_springs == '':
        total_number_of_combinations = 0
        
    # to many records for the list
    elif sum(records_list) + (len(records_list) - 1) > number_of_springs:
        total_number_of_combinations = 0
    
    # empty record list
    elif len(records_list) == 0:
        total_number_of_combinations = 1
    
    # if only one item in records
    elif len(records_list) == 1:
            total_number_of_combinations = number_of_springs - records_list[0] + 1
        
    # else if multiple items in records
    else:
        for index in range(number_of_springs):
            
            # check if current index could be start of broken springs
            if (index + records_list[0] < len(list_of_springs) 
                and list_of_springs[index + records_list[0]] != '#'):

                new_list_of_springs = list_of_springs[(records_list[0] + 1 + index):]
                new_record_list = records_list[1:]

                total_number_of_combinations += get_number_of_combinations(new_list_of_springs, new_record_list)
    
    print("combinations: " + str(total_number_of_combinations))
    
    HISTORY_DICT[(list_of_springs, str(records_list))] = total_number_of_combinations
    
    return total_number_of_combinations
```

```python
def add_points(springs_str, records_list):
    
    #print("add points to: " + str([springs_str, records_list]))
    
    # empty input
    empty_input = len(records_list) == 0 or len(springs_str) == 0
    
    # too many records
    too_many_records = (sum(records_list) + len(records_list) - 1) > len(springs_str)
    
    # if no action possible return original input
    if empty_input or too_many_records:
        return springs_str
    
    # no space for point
    if len(springs_str) == records_list[0]:
        return springs_str
    
    springs_list = list(springs_str)
    
    if springs_list[0] == '#':
        springs_list[records_list[0]] = '.'
        
    if springs_list[-1] == '#':
        springs_list[len(springs_str) - records_list[-1] - 1] = '.'
    
    #print("gives: " + "".join(springs_list))
    
    return "".join(springs_list)
```

```python
def get_number_of_combinations(springs_str, records_list):
    # this method takes an input string with with all possible symbols ('#','.','?') and the record list
    
    # remove '.' at both ends of the input string
    springs_str = springs_str.strip(".")
    
    # add points if possible
    springs_str = add_points(springs_str, records_list)
    
    # if still '.' in string, split and repeat
    if '.'in springs_str:
        
        total_number_of_combintations = 0
        
        # devide and concer by spliting the input
        springs_start_str, springs_end_str = springs_str.split('.', 1)
        
        # calculating all possible combinations by spliting record list in all possible sublists (2 lists)
        for index in range(len(records_list) + 1):
            
            start = get_number_of_combinations(springs_start_str, records_list[:index])
            end = get_number_of_combinations(springs_end_str, records_list[index:])
            
            total_number_of_combintations += start * end
            
        return total_number_of_combintations
    
    return get_combinations(springs_str, records_list)
```

```python
def solve_puzzle(puzzle_input):
    
    total_number_of_combinations = 0
    
    for line in puzzle_input.splitlines():
        
        springs_str, records_str = line.split(' ', 1)
        
        records_list = [int(record) for record in records_str.split(',')]
        
        print(get_number_of_combinations(springs_str, records_list))
        
        total_number_of_combinations += get_number_of_combinations(springs_str, records_list)
        
    return total_number_of_combinations

#solve_puzzle(puzzle_input)
```
